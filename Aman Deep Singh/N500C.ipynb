## Importing the libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv('nifty_500.csv')
dataset
```

## Conversion to Numerical

```python
convert = ["Change", "Percentage Change", "365 Day Percentage Change", "30 Day Percentage Change"]
for col in convert:
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')

dataset = pd.get_dummies(dataset, columns=["Industry", "Series"], drop_first=True)

dataset.drop(columns=["Company Name", "Symbol"], inplace=True)
dataset
```

## Confirming for missing values if any

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
imputer.fit(dataset)
dataset = pd.DataFrame(imputer.transform(dataset), columns=dataset.columns)
dataset.fillna(dataset.median(numeric_only=True), inplace=True)
total_missing_values = dataset.isnull().sum().sum()
total_missing_values
dataset
```

## Splitting the dataset into the Training set and training the model

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

dataset["Target"] = (dataset["Percentage Change"] > 0).astype(int)

X_class = dataset.drop(columns=["Last Traded Price", "Target", "Percentage Change"])
y_class = dataset["Target"]

X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, test_size=0.2, random_state=42)

svm_model = make_pipeline(StandardScaler(), SVC(kernel="rbf"))
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

lsvm_model = make_pipeline(StandardScaler(), LinearSVC(max_iter=10000))
lsvm_model.fit(X_train, y_train)
y_pred_lsvm = lsvm_model.predict(X_test)

def evaluate_model(y_true, y_pred, model_name):
    return {
        "Model": model_name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1 Score": f1_score(y_true, y_pred),
    }
svm_results = evaluate_model(y_test, y_pred_svm, "SVM")
lsvm_results = evaluate_model(y_test, y_pred_lsvm, "Linear SVM")
svm_results, lsvm_results
```
